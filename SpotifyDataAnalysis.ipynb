```python
from sklearn.neighbors import NearestNeighbors
from scipy.sparse import csr_matrix
```

```python
df2 = df[df['popularity'] >= 50]
df2['popularity'] = df2['popularity']/100
df2['time_signature'] = df2['time_signature']/5
df2['key'] = (df2['key'] - df2['key'].min())/(df2['key'].max() - df2['key'].min()) 
df2['loudness'] = (df2['loudness'] - df2['loudness'].min())/(df2['loudness'].max() - df2['loudness'].min()) 
df2['tempo'] = (df2['tempo'] - df2['tempo'].min())/(df2['tempo'].max() - df2['tempo'].min()) 
df2.describe()
```

```python
df2 = df2.drop('Unnamed: 0', axis=1)
```

```python
df2.set_index('track_id')

selected = ['popularity', 'danceability','energy','loudness' ,'key', 'speechiness', 'acousticness' ,'instrumentalness','liveness','valence','tempo']
df_knn = df2[selected]
df_knn.info()
```

```python
model_knn = NearestNeighbors(algorithm='kd_tree', n_neighbors=10)
```

```python
song_matrix = csr_matrix(df_knn.values)
```

```python
model_knn.fit(song_matrix)
```

```python
def recommend(idx, model=model_knn, n=5):
    query = df_knn.loc[idx].to_numpy().reshape(1, -1)
    distances, indices = model.kneighbors(query, n_neighbors=n)
    
    return df2.iloc[indices.flatten()]
```

```python
name = input('Enter song title: ')
filt = df2[df2['track_name'].str.contains(name, case=False) | df2['artists'].str.contains(name, case=False)]
filt[['track_name', 'album_name' , 'artists']]
```

```python
idx = int(input('Enter index: '))
print('Results: ')
recommend(idx, model_knn, 8)
```
